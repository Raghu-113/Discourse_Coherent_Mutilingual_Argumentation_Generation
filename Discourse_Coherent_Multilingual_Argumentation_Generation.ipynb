```
import torch
from transformers import MarianMTModel, MarianTokenizer, BartForConditionalGeneration, BartTokenizer
import spacy
from gramformer import Gramformer
```

```
# Initialize sentence correction model
gf = Gramformer(models=1)  # Use model 1 for grammar correction

# Load translation and generation models
models = {
    'en': {
        'tokenizer': BartTokenizer.from_pretrained('facebook/bart-large'),
        'model': BartForConditionalGeneration.from_pretrained('facebook/bart-large')
    },
    'de': {
        'tokenizer': MarianTokenizer.from_pretrained('Helsinki-NLP/opus-mt-en-de'),
        'model': MarianMTModel.from_pretrained('Helsinki-NLP/opus-mt-en-de')
    },
    'fr': {
        'tokenizer': MarianTokenizer.from_pretrained('Helsinki-NLP/opus-mt-en-fr'),
        'model': MarianMTModel.from_pretrained('Helsinki-NLP/opus-mt-en-fr')
    },
    'es': {
        'tokenizer': MarianTokenizer.from_pretrained('Helsinki-NLP/opus-mt-en-es'),
        'model': MarianMTModel.from_pretrained('Helsinki-NLP/opus-mt-en-es')
    }
}
```

```
# Load spaCy language models
nlp_models = {
    'en': spacy.load('en_core_web_sm'),
    'de': spacy.load('de_core_news_sm'),
    'fr': spacy.load('fr_core_news_sm'),
    'es': spacy.load('es_core_news_sm')
}
```

```
# Sentence correction function
def correct_sentence(text):
    corrected_sentences = list(gf.correct(text))
    return corrected_sentences[0] if corrected_sentences else text

# Argument generation using BART
def generate_argument(text, model, tokenizer):
    inputs = tokenizer.encode(text, return_tensors='pt', truncation=True)
    outputs = model.generate(inputs, max_length=500, num_beams=5, early_stopping=True)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Discourse parsing using spaCy
def parse_discourse(text, nlp):
    doc = nlp(text)
    return [sent.text for sent in doc.sents]

# Translation using MarianMT
def translate_text(text, to_lang):
    tokenizer = models[to_lang]['tokenizer']
    model = models[to_lang]['model']
    inputs = tokenizer.encode(text, return_tensors='pt', truncation=True)
    outputs = model.generate(inputs, max_length=500, num_beams=5, early_stopping=True)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```
def main():
    # Input text
    text = "Climate change are big problem, we needs action now to solve it."

    # Step 1: Correct sentence
    corrected_text = correct_sentence(text)
    print("‚úÖ Corrected Sentence:\n", corrected_text, "\n")

    # Step 2: Generate argument in English
    argument_en = generate_argument(corrected_text, models['en']['model'], models['en']['tokenizer'])
    print("‚úÖ Generated Argument in English:\n", argument_en, "\n")

    # Step 3: Translate and analyze discourse
    target_languages = ['de', 'fr', 'es']
    for lang in target_languages:
        print(f"\nüåê Language: {lang.upper()}")
        translated = translate_text(argument_en, lang)
        print(f"‚úÖ Translated Argument:\n{translated}\n")

        # Discourse analysis
        discourse_source = parse_discourse(argument_en, nlp_models['en'])
        discourse_target = parse_discourse(translated, nlp_models[lang])

        print("üß† Discourse Coherence (English):", discourse_source)
        print(f"üß† Discourse Coherence ({lang.upper()}):", discourse_target)

# Run the function
main()
```
